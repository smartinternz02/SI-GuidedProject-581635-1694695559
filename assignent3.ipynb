```python

q1 = df.body_mass_g.quantile(0.25)
q3 = df.body_mass_g.quantile(0.75)
```

```python
print(q1)
print(q3)
```

```python
IQR = q3-q1
```

```python
IQR
```

```python
upper_limit = q3+1.5*IQR
```

```python
lower_limit =q1-1.5*IQR
```

```python

df.median()
```

```python
df['body_mass_g'] = np.where(df['body_mass_g']>upper_limit,30,df['body_mass_g'])
```

```python
df['culmen_length_mm'] = np.where(df['culmen_length_mm']>upper_limit,30,df['culmen_length_mm'])
df['culmen_depth_mm'] = np.where(df['culmen_depth_mm']>upper_limit,30,df['culmen_depth_mm'])
df['flipper_length_mm'] = np.where(df['flipper_length_mm']>upper_limit,30,df['flipper_length_mm'])
```

```python
sns.boxplot(df.body_mass_g)
sns.boxplot(df.culmen_depth_mm)
sns.boxplot(df.culmen_length_mm)
sns.boxplot(df.flipper_length_mm)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df.sex = le.fit_transform(df.sex)
df.island = le.fit_transform(df.island)
df.head()
```

```python
df ['culmen_length_mm'] =df['culmen_length_mm'].fillna (df ['culmen_length_mm'].median())
df ['culmen_depth_mm']=df['culmen_depth_mm'].fillna (df ['culmen_depth_mm'].median ())
df ['flipper_length_mm'] =df['flipper_length_mm'].fillna (df['flipper_length_mm'].median ())
df ['body_mass_g'] = df ['body_mass_g'].fillna (df['body_mass_g'].median ())

df = df ['sex'].dropna ()

df
```

```python
df_main = pd.get_dummies(df,columns =['island'])
df_main.head()
```

```python

df_main.corr()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df_main.corr(),annot =True)
```
